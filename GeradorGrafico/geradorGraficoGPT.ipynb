```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

DATA_PATH_NAOBINARIO = "../ResultadoSalvo/Resultado_Completo_ClassificadorNaoBinario_GPT.csv"
DATA_PATH_BINARIO = "../ResultadoSalvo/Resultado_Completo_ClassificadorNaoBinario_GPT.csv"
```

```python
dados_NaoBinario = pd.read_csv(DATA_PATH_NAOBINARIO,index_col="indice")
dados_Binario = pd.read_csv(DATA_PATH_BINARIO,index_col="indice")

dados_Binario = dados_Binario[(dados_Binario['sentimento_GPT'] != 'x') & (dados_Binario['sentimento_humano'] !='x')]
```

Matriz de confusão do BERT não binário com 10 epocas de treinamento

```python
disp = ConfusionMatrixDisplay.from_predictions(
        y_true=dados_NaoBinario['sentimento_humano'],
        y_pred=dados_NaoBinario['sentimento_GPT'],
        cmap=plt.cm.Blues,
        display_labels=['Negatives','Positives','Neutral'],
        normalize="true",
    )
disp.ax_.set_title("Confusion Matrix")
```

Matriz de confusão do BERT não binário com 5 epocas de treinamento

Matriz de confusão do BERT binário com 10 épocas de trienamento

```python
disp = ConfusionMatrixDisplay.from_predictions(
        y_true=dados_Binario['sentimento_humano'],
        y_pred=dados_Binario['sentimento_GPT'],
        cmap=plt.cm.Blues,
        display_labels=['Negatives','Positives'],
        normalize="true",
    )
disp.ax_.set_title("Confusion Matrix")
```
