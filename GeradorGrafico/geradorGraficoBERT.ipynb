```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

DATA_PATH_NAOBINARIO_10 = "../ResultadoSalvo/Resultado_Completo_ClassificadorNaoBinario_BERT10.csv"
DATA_PATH_BINARIO_10 = "../ResultadoSalvo/Resultado_Completo_ClassificadorBinario_BERT10.csv"

DATA_PATH_NAOBINARIO_5 = "../ResultadoSalvo/Resultado_Completo_ClassificadorNaoBinario_BERT5.csv"
DATA_PATH_BINARIO_5 = "../ResultadoSalvo/Resultado_Completo_ClassificadorBinario_BERT5.csv"

```

```python
dados_NaoBinario_10 = pd.read_csv(DATA_PATH_NAOBINARIO_10,index_col="indice")
dados_Binario_10 = pd.read_csv(DATA_PATH_BINARIO_10,index_col="indice")

dados_NaoBinario_5 = pd.read_csv(DATA_PATH_NAOBINARIO_5,index_col="indice")
dados_Binario_5 = pd.read_csv(DATA_PATH_BINARIO_5,index_col="indice")
```

Matriz de confusão do BERT não binário com 10 epocas de treinamento

```python
disp = ConfusionMatrixDisplay.from_predictions(
        y_true=dados_NaoBinario_10['sentimento_humano'],
        y_pred=dados_NaoBinario_10['sentimento_BERT'],
        cmap=plt.cm.Blues,
        display_labels=['Negatives','Positives','Neutral'],
        normalize="true",
    )
disp.ax_.set_title("Confusion Matrix")
```

Matriz de confusão do BERT não binário com 5 epocas de treinamento

```python
disp = ConfusionMatrixDisplay.from_predictions(
        y_true=dados_NaoBinario_5['sentimento_humano'],
        y_pred=dados_NaoBinario_5['sentimento_BERT'],
        cmap=plt.cm.Blues,
        display_labels=['Negatives','Positives','Neutral'],
        normalize="true",
    )
disp.ax_.set_title("Confusion Matrix")
```

Matriz de confusão do BERT binário com 10 épocas de trienamento

```python
disp = ConfusionMatrixDisplay.from_predictions(
        y_true=dados_Binario_10['sentimento_humano'],
        y_pred=dados_Binario_10['sentimento_BERT'],
        cmap=plt.cm.Blues,
        display_labels=['Negatives','Positives'],
        normalize="true",
    )
disp.ax_.set_title("Confusion Matrix")
```

Matriz de confusão do BERT binário com 5 épocas de trienamento

```python
disp = ConfusionMatrixDisplay.from_predictions(
        y_true=dados_Binario_5['sentimento_humano'],
        y_pred=dados_Binario_5['sentimento_BERT'],
        cmap=plt.cm.Blues,
        display_labels=['Negatives','Positives'],
        normalize="true",
    )
disp.ax_.set_title("Confusion Matrix")
```
